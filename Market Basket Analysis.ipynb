# Association rule || Market Basket Analyses

- Metrics/ rule:
         1. Support: (presence of 1 item in transactions i.e total tranx 5, item 1 presence in 3, support 3/5
         2. Confidence: (combination of two item same transaction, 3 times purchase for item 1, 2 times for
         item 2, confidence will be 2/3
         3. lift: confidence(AB)/Support(B)
         
- Aprior algorithm

```python
# Loading neccesary packages
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
# Reading Data 

# Define the file path
file_path = '/Users/ridwanahmed/Desktop/Porfolio/Python Project/Datasets/Online Retail.xlsx'

# Read the Excel file into a pandas DataFrame
df = pd.read_excel(file_path)

# Call the DataFrame
df
```

```python
df.head()
```

## DATA Preparation

```python
#Data Cleaning

df['Description'] = df['Description'].str.strip() #removes spaces from beginning and end
df.dropna(axis=0, subset=['InvoiceNo'], inplace=True) #removes duplicate invoice
df['InvoiceNo']= df['InvoiceNo'].astype('str') #converting invoice number to be string
df = df[~df['InvoiceNo'].str.contains('C')] #remove the credit transactions 

df.head()
```

```python
df.shape
```

```python
# Value or record count for different countries

df['Country'].value_counts()
```

```python
# Separating transactions for Germany
# filtering data on one country, do not want to run the analysis for all 550k tranx
# making basket of tranx

mybasket = (df[df['Country'] == "Germany"]
           .groupby(['InvoiceNo', 'Description'])['Quantity']
           .sum().unstack().reset_index().fillna(0)
           .set_index('InvoiceNo'))
```

```python
mybasket
```

```python
# Converting all positive values to 1 and everything else to 0
# I want df to become only 0 and 1 because that is what my algorithm of association analysis expects the input

def my_encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
    
my_basket_sets = mybasket.applymap(my_encode_units)
my_basket_sets.drop('POSTAGE', inplace= True, axis=1) #remove 'postage' as an item
```

## Training Model

```python
# Generating frequent itemsets
# using aprior algorithm

my_frequent_itemsets = apriori(my_basket_sets, min_support=0.07, use_colnames=True)
```

```python
# Generating rules

my_rules = association_rules(my_frequent_itemsets, metric="lift", min_threshold=1)
```

```python
# Viewing top 100 rules
# Support, Lift and Confidence
# Know about antecedent, support, consequent and antecedent support

my_rules.head(100)
```
